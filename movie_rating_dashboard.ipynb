```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("F:/ProgrammingJavaPythonHtml/Year 2025/archive/movie.csv")
rating = pd.read_csv("F:/ProgrammingJavaPythonHtml/Year 2025/archive/rating.csv")
```

```python
df.head(50)
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates(subset='title')
```

```python
np.mean(rating['rating'])
```

```python
np.max(rating['rating'])
```

```python
np.min(rating['rating'])
```

```python
genre_counts = df['genres'].value_counts()
genre_counts
```

```python
rating.head(50)
```

```python
merged_data = pd.merge(rating, df, on="movieId")
```

```python
merged_data
```

```python
merged_data.isnull().sum()
```

```python
merged_data['release_year'] = merged_data['title'].str.extract(r"\((\d{4})\)").astype(float)
```

```python
top_movies = merged_data.groupby('title')['rating'].mean().sort_values(ascending=False).head(10)
top_movies
```

```python
genre_counts = merged_data['genres'].value_counts()
genre_counts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top_rated_movies = merged_data.groupby('title')['rating'].mean().sort_values(ascending=False).head(10)

# Bar chart
top_rated_movies.plot(kind='bar', color='green')
plt.title("Top 10 Movies by Average Rating")
plt.xlabel("Movie")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter movies by genre
genre = "Action"  # Example genre
genre_filtered = merged_data[merged_data['genres'].str.contains(genre, na=False)]
genre_filtered
```

```python
most_rated_movies = merged_data.groupby('title').size().sort_values(ascending=False).head(10)
most_rated_movies
```

```python
merged_data.to_csv("cleaned_movie_data.csv", index=False)
```

```python
import streamlit as st

# Load the merged data
data = pd.read_csv("cleaned_movie_data.csv")

st.title("Movie Ratings Dashboard")

# Filter by Genre
genre = st.selectbox("Select Genre", data['genres'].unique())
filtered_data = data[data['genres'].str.contains(genre, na=False)]

st.write(f"Top Movies in {genre}")
st.dataframe(filtered_data[['title', 'rating']].sort_values(by='rating', ascending=False).head(10))
```
